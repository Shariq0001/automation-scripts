# 🔍 CAT Data Lake — Schema Validator

**Purpose:** Silver layer (Trino) se fetch kiya hua data, CAT event schema CSV ke against validate karna.

| Source | Detail |
|--------|--------|
| **Schema** | CAT event schema CSV (`eventtype`, `positionkey`, `fieldname`, `data_type`) |
| **Data** | Silver layer → Trino query (WHERE clause se filter) |

### Checks Performed
| # | Check | Description |
|---|-------|-------------|
| 1 | **Column Count** | Schema unique columns vs Trino columns — missing/extra list |
| 2 | **Data Type Validation** | Har schema column ki values datatype se match karti hain? |
| 3 | **NULL / Garbage / Type Error** | Per-column breakdown with null% |
| 4 | **Pipeline Columns Check** | `boothid`, `ingestedfordate` etc. — ye null nahi hone chahiye |

### Notes
- Same fieldname multiple events mein hoga (e.g. `price` MEOR aur MEOC dono mein) — **union schema** use hogi, column ek hi baar validate hoga
- Pipeline-injected columns schema mein nahi hote — inhe alag section mein validate kiya jata hai
- NULL values ko **FAIL nahi** kiya jata (kyunki multi-event file mein nulls expected hain) — sirf report hota hai
- Pipeline columns null hone par **❌ FAIL** hoga

### Supported Data Types
| Schema Type | Validates As |
|-------------|-------------|
| `String` | Any non-empty text |
| `Long` | Whole number (no decimals) |
| `Boolean` | true/false/1/0/yes/no/y/n/t/f |
| `Decimal` | High-precision numeric |
| `Date` | Parseable date string |
| `Timestamp` | Parseable datetime string |
| `Enum` | Treated as String |

## 🔧 Cell 3 — Helper Functions

```python
def normalize_name(value):
    """Column names lowercase + spaces underscore."""
    value = str(value).strip()
    value = re.sub(r'\s+', '_', value)
    return value.lower()


def normalize_df_columns(df):
    df = df.copy()
    df.columns = [normalize_name(col) for col in df.columns]
    return df


def normalize_dtype(raw):
    """Schema data_type string ko standard mapped type mein convert karo."""
    if pd.isna(raw):
        return None
    cleaned = str(raw).strip().lower()

    if 'decimal' in cleaned or 'numeric' in cleaned or 'number' in cleaned:
        return 'decimal'

    base = re.sub(r'\(.*?\)', '', cleaned).strip()

    aliases = {
        'long'     : 'bigint',
        'bigint'   : 'bigint',
        'int'      : 'bigint',
        'integer'  : 'bigint',
        'string'   : 'string',
        'str'      : 'string',
        'varchar'  : 'string',
        'text'     : 'string',
        'enum'     : 'string',   # Enum = String ki tarah treat
        'boolean'  : 'boolean',
        'bool'     : 'boolean',
        'date'     : 'date',
        'timestamp': 'timestamp',
        'datetime' : 'timestamp',
        'double'   : 'double',
        'float'    : 'double',
    }
    return aliases.get(base)


def is_null_like(value):
    if pd.isna(value):
        return True
    s = str(value).strip()
    return s == '' or s.lower() in {'null', 'none', 'nan', 'nat'}


def validate_value_detailed(value, mapped_type):
    """
    Ek value validate karo.
    Returns: (status_str, is_error_bool)
      status: 'NULL' | 'VALID' | 'TYPE_ERROR' | 'GARBAGE' | 'SKIPPED'
    """
    if is_null_like(value):
        return 'NULL', False

    s = str(value).strip()

    try:
        if mapped_type == 'string':
            return 'VALID', False

        if mapped_type == 'bigint':
            if re.fullmatch(r'[+-]?\d+', s):
                return 'VALID', False
            if re.fullmatch(r'[+-]?\d+\.0+', s):   # 3.0 = acceptable
                return 'VALID', False
            if re.fullmatch(r'[+-]?\d+\.\d+', s):  # 3.5 = type error
                return 'TYPE_ERROR', True
            return 'GARBAGE', True

        if mapped_type == 'double':
            float(s)
            return 'VALID', False

        if mapped_type == 'decimal':
            Decimal(s)
            return 'VALID', False

        if mapped_type == 'boolean':
            if s.lower() in {'true', 'false', '1', '0', 'yes', 'no', 'y', 'n', 't', 'f'}:
                return 'VALID', False
            return 'TYPE_ERROR', True

        if mapped_type == 'date':
            parsed = pd.to_datetime(s, errors='coerce')
            if pd.isna(parsed):
                return 'GARBAGE', True
            return 'VALID', False

        if mapped_type == 'timestamp':
            parsed = pd.to_datetime(s, errors='coerce')
            if pd.isna(parsed):
                return 'GARBAGE', True
            return 'VALID', False

        return 'SKIPPED', False

    except InvalidOperation:
        return 'TYPE_ERROR', True
    except Exception:
        return 'GARBAGE', True


print('✅ Helper functions ready!')
```

## 📚 Cell 4 — Schema CSV Load karo

```python
def load_schema(schema_file, event_filter=None):
    """
    CAT schema CSV load karo.
    - event_filter: list of event types ya None (all)
    - Returns union schema (unique fieldnames across all events)
      kyunki same field (e.g. price) multiple events mein hoga
    """
    # CSV ya Excel dono support
    path_lower = schema_file.lower()
    if path_lower.endswith('.csv'):
        schema_df = pd.read_csv(schema_file)
    elif path_lower.endswith(('.xlsx', '.xls')):
        schema_df = pd.read_excel(schema_file)
    else:
        raise ValueError(f'Unsupported file format: {schema_file}')

    # Normalize column names
    schema_df = normalize_df_columns(schema_df)

    required = {'eventtype', 'positionkey', 'fieldname', 'data_type'}
    missing_cols = required - set(schema_df.columns)
    if missing_cols:
        raise ValueError(f'Schema file mein ye columns nahi hain: {missing_cols}')

    schema_df['eventtype']       = schema_df['eventtype'].astype(str).str.strip()
    schema_df['fieldname']       = schema_df['fieldname'].astype(str).str.strip()
    schema_df['fieldname_norm']  = schema_df['fieldname'].map(normalize_name)
    schema_df['positionkey_num'] = pd.to_numeric(schema_df['positionkey'], errors='coerce')

    # Event filter apply karo
    if event_filter:
        event_filter_set = {str(x).strip() for x in event_filter}
        schema_df = schema_df[schema_df['eventtype'].isin(event_filter_set)].copy()
        print(f'🔍 Event filter applied: {event_filter_set}')

    schema_df = schema_df.sort_values(
        ['eventtype', 'positionkey_num', 'fieldname_norm'], na_position='last'
    ).reset_index(drop=True)

    # Union schema — same fieldname across events = ek hi entry (first occurrence)
    union_schema = (
        schema_df[['fieldname_norm', 'data_type']]
        .drop_duplicates(subset=['fieldname_norm'], keep='first')
        .copy()
    )

    field_type_map = dict(zip(union_schema['fieldname_norm'], union_schema['data_type']))

    print(f'\n📄 Schema rows loaded       : {len(schema_df)}')
    print(f'🗂️  Unique event types       : {schema_df["eventtype"].nunique()} → {sorted(schema_df["eventtype"].unique().tolist())}')
    print(f'🔢 Unique schema columns    : {len(union_schema)}')
    print(f'ℹ️  Note: Fields appearing in multiple events use union schema (one entry per fieldname)')

    return schema_df, union_schema, field_type_map


schema_df, union_schema, field_type_map = load_schema(SCHEMA_FILE, SCHEMA_EVENT_FILTER)
```

## 🔌 Cell 5 — Trino se Data Fetch karo

```python
def get_data_from_trino():
    auth = BasicAuthentication(TRINO_CONFIG['user'], TRINO_CONFIG['password'])
    conn = connect(
        host        = TRINO_CONFIG['host'],
        port        = TRINO_CONFIG['port'],
        user        = TRINO_CONFIG['user'],
        auth        = auth,
        http_scheme = 'https',
        verify      = False,
        catalog     = TRINO_CONFIG['catalog'],
        schema      = TRINO_CONFIG['schema']
    )

    query = f"SELECT * FROM {TARGET_TABLE} WHERE {WHERE_CLAUSE}"
    if FETCH_LIMIT is not None:
        query += f" LIMIT {int(FETCH_LIMIT)}"

    print('⏳ Trino se data fetch ho raha hai...')
    print(f'   Query: {query[:120]}...' if len(query) > 120 else f'   Query: {query}')

    cur = conn.cursor()
    cur.execute(query)
    rows = cur.fetchall()
    cols = [desc[0] for desc in cur.description]

    df = pd.DataFrame(rows, columns=cols)
    df = normalize_df_columns(df)

    print(f'\n✅ Data fetch ho gaya!')
    print(f'   📊 Total rows    : {len(df)}')
    print(f'   📋 Total columns : {len(df.columns)}')
    return df


data_df = get_data_from_trino()
```

## 🔢 Cell 6 — Column Count Check (Schema vs Trino)

```python
# Schema columns (113) + pipeline columns (8) = expected total (121)
schema_columns     = set(union_schema['fieldname_norm'].tolist())
pipeline_cols_norm = set(normalize_name(c) for c in PIPELINE_COLUMNS)
expected_columns   = schema_columns | pipeline_cols_norm   # 113 + 8 = 121
data_columns       = set(data_df.columns.tolist())

matched_cols       = expected_columns & data_columns       # all 121 expected cols jo Trino mein hain
schema_matched     = schema_columns & data_columns         # sirf schema cols (type validation k liye)
missing_from_query = expected_columns - data_columns       # expected mein hai, Trino mein nahi
extra_in_query     = data_columns - expected_columns       # Trino mein hai, expected mein nahi

print('\n================ COLUMN COUNT CHECK ================\n')
print(f'📄 Schema columns                    : {len(schema_columns)}')
print(f'🔧 Pipeline columns                  : {len(pipeline_cols_norm)}')
print(f'📊 Expected total (schema + pipeline): {len(expected_columns)}')
print(f'🗄️  Trino columns (actual)            : {len(data_columns)}')
print(f'✅ Matched                           : {len(matched_cols)}')

print(f'\n🔴 Missing from Trino ({len(missing_from_query)}) — expected mein hai, data mein nahi:')
if missing_from_query:
    for i, col in enumerate(sorted(missing_from_query), 1):
        print(f'  {i}. {col}')
else:
    print('  ✅ None — sab expected columns Trino mein hain')

print(f'\n🔵 Extra in Trino ({len(extra_in_query)}) — Trino mein hai, expected mein nahi:')
if extra_in_query:
    for i, col in enumerate(sorted(extra_in_query), 1):
        print(f'  {i}. {col}')
else:
    print('  ✅ None')

column_compare_df = pd.DataFrame([
    {'column': col, 'status': 'MISSING_FROM_TRINO'} for col in sorted(missing_from_query)
] + [
    {'column': col, 'status': 'EXTRA_IN_TRINO'} for col in sorted(extra_in_query)
] + [
    {'column': col, 'status': 'MATCHED'} for col in sorted(matched_cols)
])
```

## ✅ Cell 7 — Schema Column Datatype Validation

```python
def validate_all_columns(data_df, field_type_map, schema_matched, pipeline_cols_norm):
    """
    Schema columns + pipeline columns — ek hi loop, ek hi table.
    - Schema cols: full type validation (NULL report only, type error = FAIL)
    - Pipeline cols: sirf null check — null hone par FAIL, [PIPELINE] tag se highlight
    """
    results = []
    error_rows_list = []

    # Schema cols first (sorted), phir pipeline cols
    schema_cols_sorted   = sorted(schema_matched)
    pipeline_cols_sorted = sorted(pipeline_cols_norm)
    all_cols_ordered     = schema_cols_sorted + pipeline_cols_sorted
    total                = len(all_cols_ordered)

    print(f'⏳ Validating {total} columns ({len(schema_cols_sorted)} schema + {len(pipeline_cols_sorted)} pipeline)...\n')

    for col in all_cols_ordered:
        is_pipeline = col in pipeline_cols_norm

        if col not in data_df.columns:
            tag = ' [PIPELINE]' if is_pipeline else ''
            print(f'  {(col + tag):<52} | {"MISSING":<12} | {"—":>6} | {"—":>7} | {"—":>8} | ❌ MISSING FROM TRINO')
            results.append({'column': col, 'schema_type': 'PIPELINE' if is_pipeline else 'SCHEMA',
                            'mapped_type': '-', 'row_count': 0, 'null_count': 0,
                            'null_pct': 0, 'type_error_count': 0, 'garbage_count': 0,
                            'status': 'MISSING', 'is_pipeline': is_pipeline})
            continue

        values    = data_df[col].tolist()
        row_count = len(values)

        null_count       = 0
        type_error_count = 0
        garbage_count    = 0

        if is_pipeline:
            # Pipeline: sirf null check
            null_count = sum(1 for v in values if is_null_like(v))
            raw_type   = 'PIPELINE'
            mapped_type = 'pipeline'
        else:
            # Schema: full type validation
            raw_type    = field_type_map.get(col, 'unknown')
            mapped_type = normalize_dtype(raw_type)
            for row_idx, val in enumerate(values):
                status, is_error = validate_value_detailed(val, mapped_type)
                if status == 'NULL':
                    null_count += 1
                elif status == 'TYPE_ERROR':
                    type_error_count += 1
                    error_rows_list.append({'column': col, 'schema_type': raw_type,
                                           'row_index': row_idx, 'value': str(val)[:100], 'error': 'TYPE_ERROR'})
                elif status == 'GARBAGE':
                    garbage_count += 1
                    error_rows_list.append({'column': col, 'schema_type': raw_type,
                                           'row_index': row_idx, 'value': str(val)[:100], 'error': 'GARBAGE'})

        null_pct   = round((null_count / row_count) * 100, 2) if row_count > 0 else 0
        has_errors = (type_error_count > 0 or garbage_count > 0)

        if is_pipeline:
            status_label = '❌ FAIL — NULL FOUND' if null_count > 0 else 'PASS'
            has_errors   = null_count > 0   # pipeline mein null = fail
        else:
            if has_errors:
                status_label = 'FAIL'
            elif null_count > 0:
                status_label = 'PASS | NULLS_PRESENT'
            else:
                status_label = 'PASS'

        results.append({
            'column'          : col,
            'schema_type'     : raw_type,
            'mapped_type'     : mapped_type,
            'row_count'       : row_count,
            'null_count'      : null_count,
            'null_pct'        : null_pct,
            'type_error_count': type_error_count,
            'garbage_count'   : garbage_count,
            'status'          : status_label,
            'is_pipeline'     : is_pipeline
        })

        # ── Print row ──
        null_flag   = f'⚠️  HIGH NULL ({null_pct:.0f}%)' if null_pct >= 50 else ('' if null_count == 0 else f'({null_pct:.0f}% null)')
        status_icon = '❌ FAIL' if has_errors else '✅ PASS'
        tag         = ' *[PIPELINE]*' if is_pipeline else ''
        col_display = col + tag
        print(f'  {col_display:<52} | {str(raw_type):<12} | {null_count:>6} | {garbage_count:>7} | {type_error_count:>8} | {null_flag:<22} {status_icon}')

    print('\n' + '='*110)
    print('✅ Validation complete')
    return pd.DataFrame(results), pd.DataFrame(error_rows_list)


# ── Header ──
print('=' * 110)
print('🔍 CAT DATA LAKE: DETAILED VALIDATION REPORT')
print('   * [PIPELINE] = pipeline-injected columns — null nahi hone chahiye')
print('=' * 110)
print(f'  {"COLUMN":<52} | {"SCHEMA TYPE":<12} | {"NULLS":>6} | {"GARBAGE":>7} | {"TYPE ERR":>8} | STATUS')
print('-' * 110)
validation_summary_df, error_df = validate_all_columns(data_df, field_type_map, schema_matched, pipeline_cols_norm)
```

## 📊 Cell 9 — Full Summary Report

```python
failed_schema_cols  = validation_summary_df[validation_summary_df['status'] == 'FAIL'] if len(validation_summary_df) else pd.DataFrame()
nulls_present_cols  = validation_summary_df[validation_summary_df['null_count'].fillna(0) > 0] if len(validation_summary_df) else pd.DataFrame()

final_summary = {
    'schema_unique_columns' : len(schema_columns),
    'trino_total_columns'   : len(data_columns),
    'pipeline_columns'      : len(pipeline_cols_norm),
    'matched_columns'       : len(matched_cols),
    'missing_from_trino'    : len(missing_from_query),
    'extra_in_trino'        : len(extra_in_query),
    'validated_columns'     : len(validation_summary_df),
    'failed_columns'        : len(failed_schema_cols),
    'columns_with_nulls'    : len(nulls_present_cols),
    'total_error_rows'      : len(error_df),
}

summary_df = pd.DataFrame(list(final_summary.items()), columns=['metric', 'value'])

print('\n================ FINAL SUMMARY ================\n')
display(summary_df)

# Schema validation table
print('\n📋 Schema Column Validation (all columns):')
display(validation_summary_df[['column','schema_type','mapped_type','row_count','null_count','null_pct','type_error_count','garbage_count','status']])

# Failed schema columns
if not failed_schema_cols.empty:
    print(f'\n❌ Failed Schema Columns ({len(failed_schema_cols)}):')
    display(failed_schema_cols[['column','schema_type','type_error_count','garbage_count','status']])

# Pipeline columns

# Null report
null_only_df = nulls_present_cols[['column','schema_type','row_count','null_count','null_pct','status']].copy()
null_only_df = null_only_df.sort_values(['null_pct','null_count'], ascending=[False, False])
print(f'\n⚠️  Columns with NULLs ({len(null_only_df)}) — sorted by null%:')
display(null_only_df)
```
